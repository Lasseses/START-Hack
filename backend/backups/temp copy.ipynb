```python
import time

# Example criteria and example values to test
criteria_examples = {
    "revenue": "greater than 1000000",
    "net_income": "is positive",
    "EBITDA": "is positive",
    "operating_income": "greater than 0",
    "EPS": "greater than 1.5",
    "dividend_yield": "greater than 2%",
    "PE_ratio": "less than 20",
    "market_cap": "greater than 1000000000",
    "price": "greater than 50",
    "employees": "more than 10000",
    "debt_to_equity": "less than 1",
    "return_on_equity": "greater than 15%",
    "operating_margin": "greater than 10%",
    "profit_margin": "greater than 20%",
    "revenue_growth": "is positive",
    "net_income_growth": "is positive",
    "cash_flow": "greater than 500000",
    "free_cash_flow": "greater than 1000000",
    "total_assets": "greater than 100000000",
    "total_liabilities": "less than total_assets",
    "current_ratio": "greater than 1",
    "quick_ratio": "greater than 0.8",
    "price_to_book": "less than 3",
    "beta": "greater than 1",
    "sector": "equals 'Technology'",
    "industry": "equals 'Banking'",
    "country": "equals 'USA'",
    "founded_year": "after 1950",
    "exchange": "equals 'NASDAQ'",
    "short_interest": "greater than 5%",
    "dividend_payout_ratio": "greater than 30%",
    "insider_ownership": "greater than 5%",
    "institutional_ownership": "greater than 70%",
    "gross_margin": "greater than 40%",
    "capex": "greater than 100000",
    "R&D_expense": "greater than 500000",
    "EPS_growth": "greater than 10%",
    "price_target": "less than 1.5 times current price"
}

working_criteria = []

# Loop over each criterion and test it
for criterion, example_value in criteria_examples.items():
    # Build the query string in the desired JSON format
    query = f'{{"{criterion}": "{example_value}"}}'
    
    try:
        # Call the API function with the query.
        # call_searchwithcriteria is assumed to be defined elsewhere.
        result = call_searchwithcriteria(query)
        convert_searchwithcriteria(query, result)
        # If no error occurs, add the criterion to the working list.
        working_criteria.append(criterion)
        print(f"Criterion '{criterion}' worked. Result: {result}")
    except Exception as e:
        print(f"Error with criterion '{criterion}': {e}")

    time.sleep(1)

print("Working criteria:", working_criteria)
```

# companydatasearch

```python
def companydatasearch(query:str):
    """
    This tool is useful when you need information about one or more companies, such as employee numbers, 
    market or financial information, ratios, fundamentals, etc. Input the query in the following format: 
    '{"company1": "information to retrieve|yyyyQq"; "company2": "information to retrieve|yyyy"}'.
    Example: '{"Caixabank": "number of employees|2023"; "BBVA": "number of employees|2023"}'.
    """
    url = ("https://idchat-api-containerapp01-dev.orangepebble-16234c4b."
           "switzerlandnorth.azurecontainerapps.io//companydatasearch"
           f"?query={query}")
    response = requests.post(url)
    return response.json()
```
